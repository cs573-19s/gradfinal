```python
import pandas

data = pandas.read_csv("storms.csv", low_memory=False)
```

```python
data
```

```python
data['Nature'].value_counts()
```

```python
TS = data.loc[data['Nature'] == " TS"]
TS = TS.reset_index(drop=True)
TS
```

```python
SpecificStorms = TS[TS['Serial_Num'].isin(['2005236N23285', '2005275N19274', '2004258N16300', '1998295N12284',
                                          '1994313N12278', '1979238N12324', '1974258N15295', '1963270N08327', '1900239N15318',
                                          '1935292N12281'])]

SpecificStorms
```

```python
Filtered = SpecificStorms[['Serial_Num', 'Season', 'Name', 'ISO_time', 'Latitude', 'Longitude', 'Wind(WMO)']]
Filtered = Filtered.reset_index(drop=True)
Filtered
```

```python
Filtered['Deaths'] = 0
Filtered
```

```python
for x in range(len(Filtered)):
    if(Filtered['Serial_Num'][x] == '2005275N19274'):
        Filtered['Deaths'][x] = 1668
    elif(Filtered['Serial_Num'][x] == '2005236N23285'):
        Filtered['Deaths'][x] = 1836
    elif(Filtered['Serial_Num'][x] == '2004258N16300'):
        Filtered['Deaths'][x] = 3037
    elif(Filtered['Serial_Num'][x] == '1998295N12284'):
        Filtered['Deaths'][x] = 15000
    elif(Filtered['Serial_Num'][x] == '1994313N12278'):
        Filtered['Deaths'][x] = 1152
    elif(Filtered['Serial_Num'][x] == '1979238N12324'):
        Filtered['Deaths'][x] = 2068
    elif(Filtered['Serial_Num'][x] == '1974258N15295'):
        Filtered['Deaths'][x] = 8210
    elif(Filtered['Serial_Num'][x] == '1963270N08327'):
        Filtered['Deaths'][x] = 7193
    elif(Filtered['Serial_Num'][x] == '1900239N15318'):
        Filtered['Deaths'][x] = 10000
    elif(Filtered['Serial_Num'][x] == '1935292N12281'):
        Filtered['Deaths'][x] = 2150

Filtered
```

```python
Filtered.rename(columns={'Wind(WMO)': 'Wind', 'Serial_Num': 'SerialNum', 'ISO_time': 'ISOTime'}, inplace=True)
Filtered
```

```python
Filtered.to_csv (r'FilteredStorms.csv', index = None, header=True)
```
