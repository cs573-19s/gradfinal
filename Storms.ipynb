```python
import pandas
```

```python
data = pandas.read_csv("storms.csv", low_memory=False)
data
```

```python
data['Nature'].value_counts()
```

```python
TS = data.loc[data['Nature'] == " TS"]
TS = TS.reset_index(drop=True)
TS
```

```python
Filtered = TS[['Serial_Num', 'Season', 'Name', 'ISO_time', 'Latitude', 'Longitude', 'Wind(WMO)']]
Filtered = Filtered.reset_index(drop=True)
Filtered
```

```python
Filtered.rename(columns={'Wind(WMO)': 'Wind', 'Serial_Num': 'SerialNum', 'ISO_time': 'ISOTime'}, inplace=True)
Filtered
```

```python
deaths = pandas.read_csv("casualties.csv", low_memory=False)
deaths
```

```python
merged = pandas.merge(Filtered, deaths, on=['SerialNum'])
merged
```

```python
print(merged['SerialNum'].value_counts()[50:100])
```

```python
merged.to_csv (r'FilteredStorms.csv', index = None, header=True)
```
